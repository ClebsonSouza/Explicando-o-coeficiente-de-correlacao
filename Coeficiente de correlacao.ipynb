O coeficiente de correlao, comumente representado por "r"
mede o grau de relao linear entre duas variveis.

r=1: Correlao perfeitamente positiva.

r=1: Correlao perfeitamente negativa.

r=0: Ausncia de correlao.





```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dados para o DataFrame
dados = {
    'ano': [2019, 2020, 2021, 2022, 2023, 2024],
    'consumo': [2300, 2400, 2500, 2950, 3150, 3200],
    'salrio': [2900, 3200, 3500, 3700, 3800, 4000]
}

# Criando o DataFrame
df = pd.DataFrame(dados)

display(df)
```

```python
# GRFICO DE LINHAS #
plt.figure(figsize=(8, 4))


plt.plot(df['ano'], df['consumo'], marker='o', label='Consumo', color='red')
plt.plot(df['ano'], df['salrio'], marker='x', label='Salrio', color='green')


plt.xlabel('Ano')
plt.ylabel('Valor')
plt.title('Consumo e Salrio')
plt.legend()

plt.grid(False)
plt.show()
```

```python
# GRFICO DE DISPERSO #

coeficientes = np.polyfit(df['consumo'], df['salrio'], 1)
linha_tendencia = np.poly1d(coeficientes)

plt.scatter(df['consumo'], df['salrio'], color='green', label='Dados')

plt.plot(df['consumo'], linha_tendencia(df['consumo']), color='blue', label='Linha de Tendncia')

plt.xlabel('Consumo')
plt.ylabel('Salrio')
plt.title('Grfico de Disperso: Consumo vs. Salrio')
plt.legend()

plt.grid(False)
plt.show()
```

```python
# Calculando o coeficiente de correlao
coeficiente_correlacao = round(df['consumo'].corr(df['salrio']), 2)

print("Coeficiente de correlao entre consumo e salrio:", coeficiente_correlacao)
```

Ou seja, 94% das variaes no consumo so explicadas por variaes na renda!
