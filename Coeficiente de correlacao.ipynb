O coeficiente de correlação, comumente representado por "r"
mede o grau de relação linear entre duas variáveis.

r=1: Correlação perfeitamente positiva.

r=−1: Correlação perfeitamente negativa.

r=0: Ausência de correlação.





```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dados para o DataFrame
dados = {
    'ano': [2019, 2020, 2021, 2022, 2023, 2024],
    'consumo': [2300, 2400, 2500, 2950, 3150, 3200],
    'salário': [2900, 3200, 3500, 3700, 3800, 4000]
}

# Criando o DataFrame
df = pd.DataFrame(dados)

display(df)
```

```python
# GRÁFICO DE LINHAS #
plt.figure(figsize=(8, 4))


plt.plot(df['ano'], df['consumo'], marker='o', label='Consumo', color='red')
plt.plot(df['ano'], df['salário'], marker='x', label='Salário', color='green')


plt.xlabel('Ano')
plt.ylabel('Valor')
plt.title('Consumo e Salário')
plt.legend()

plt.grid(False)
plt.show()
```

```python
# GRÁFICO DE DISPERSÃO #

coeficientes = np.polyfit(df['consumo'], df['salário'], 1)
linha_tendencia = np.poly1d(coeficientes)

plt.scatter(df['consumo'], df['salário'], color='green', label='Dados')

plt.plot(df['consumo'], linha_tendencia(df['consumo']), color='blue', label='Linha de Tendência')

plt.xlabel('Consumo')
plt.ylabel('Salário')
plt.title('Gráfico de Dispersão: Consumo vs. Salário')
plt.legend()

plt.grid(False)
plt.show()
```

```python
# Calculando o coeficiente de correlação
coeficiente_correlacao = round(df['consumo'].corr(df['salário']), 2)

print("Coeficiente de correlação entre consumo e salário:", coeficiente_correlacao)
```

Ou seja, 94% das variações no consumo são explicadas por variações na renda!
